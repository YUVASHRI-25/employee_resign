```python
df['Gender']=df['Gender'].map({'Female':int(0),'Male':int(1)})
```

```python
df["Education_Level"] = df["Education_Level"].map({
    "High School": 0,
    "Master": 1,
    "PhD": 2
})
```

```python
print("Number of unique job roles:", df["Job_Title"].nunique())
print(df["Job_Title"].unique())

```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
encoder=LabelEncoder()
df['Department']=encoder.fit_transform(df['Department'])
df['Job_Title']=encoder.fit_transform(df['Job_Title'])
```

```python
import pandas as pd

df = df.drop(columns=["Employee_ID","Hire_Date"])

# Save updated dataset
df.to_csv("Data_Updated.csv", index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
sns.countplot(x='Resigned', data=df, palette='Set2')

plt.title('Distribution of Employee Resignations')
plt.xlabel('Resigned (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.xticks([0, 1], ['No', 'Yes'], rotation=0)  

plt.show()
```
