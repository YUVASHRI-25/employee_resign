```python
import pandas as pd
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

# 1. Load dataset
df = pd.read_csv("E:/PYTHON/ML_PROJECT/new_random_5000_rows.csv")

# 2. Encode categorical variables (manual mapping)
df['Department'] = df['Department'].map({
    'Sales': 0,
    'Marketing': 1,
    'Operations': 2,
    'Customer Support': 3,
    'Legal': 4,
    'Finance': 5,
    'IT': 6,
    'HR': 7,
    'Engineering': 8
})

df['Gender'] = df['Gender'].map({
    'Other': 0,
    'Male': 1,
    'Female': 2
})

df['Job_Title'] = df['Job_Title'].map({
    'Manager': 0,
    'Engineer': 1,
    'Analyst': 2,
    'Technician': 3,
    'Specialist': 4,
    'Developer': 5,
    'Consultant': 6
})

df['Education_Level'] = df['Education_Level'].map({
    'High School': 0,
    'Bachelor': 1,
    'Master': 2,
    'PhD': 3
})

# 3. Define X and y AFTER mapping
X = df.drop(columns=['Employee_ID', 'Hire_Date', 'Resigned'])
y = df['Resigned'].map({True: 1, False: 0})

# 4. Train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# 5. Apply SMOTE
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X_train, y_train)

print("Before SMOTE:\n", y_train.value_counts())
print("\nAfter SMOTE:\n", y_res.value_counts())
```

```python
import pandas as pd

# Combine resampled features and target
smote_df = pd.concat([X_res, y_res], axis=1)

# Save to CSV
smote_df.to_csv("E:/PYTHON/ML_PROJECT/smote_resampled.csv", index=False)

```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from imblearn.over_sampling import SMOTE
x_train, x_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42)
models = {
    "Logistic Regression": LogisticRegression(max_iter=2000),  # increased iterations
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "AdaBoost": AdaBoostClassifier(),
}
```

```python
for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)

    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, zero_division=0)
    prec = precision_score(y_test, y_pred, zero_division=0)
    rec = recall_score(y_test, y_pred, zero_division=0)

    print(f"\nModel: {name}")
    print(f"  Accuracy : {acc:.5f}")
    print(f"  F1 Score : {f1:.4f}")
    print(f"  Precision: {prec:.4f}")
    print(f"  Recall   : {rec:.4f}")
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
y_prob = models["Random Forest"].predict_proba(x_test)[:, 1]
fpr, tpr, _ = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)
plt.figure(figsize=(7, 5))
plt.plot(fpr, tpr, color='blue', lw=2, label=f"Random Forest (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Random Forest")
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
y_pred = models["Random Forest"].predict(x_test)
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["No", "Yes"])
disp.plot(cmap="Blues")
plt.title("Confusion Matrix - Random Forest")
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
rf_params = {
    "n_estimators": [100, 200],
    "max_depth": [None, 10, 20],
    "min_samples_split": [2, 5],
    "min_samples_leaf": [1, 2]
}

rf = RandomForestClassifier(random_state=42)
rf_grid = GridSearchCV(rf, rf_params, cv=cv, scoring="f1", n_jobs=-1, verbose=1)
rf_grid.fit(X_res, y_res)

print("\nBest Random Forest Params:", rf_grid.best_params_)
y_pred_rf = rf_grid.predict(x_test)

print("\nRandom Forest Results:")
print("Accuracy :", accuracy_score(y_test, y_pred_rf))
print("Precision:", precision_score(y_test, y_pred_rf))
print("Recall   :", recall_score(y_test, y_pred_rf))
print("F1 Score :", f1_score(y_test, y_pred_rf))
```

```python
results = []
for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)

    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, zero_division=0)
    prec = precision_score(y_test, y_pred, zero_division=0)
    rec = recall_score(y_test, y_pred, zero_division=0)
    results.append([name, acc, f1, prec, rec])

metrics_df = pd.DataFrame(results, columns=["Model", "Accuracy", "F1 Score", "Precision", "Recall"])
display(metrics_df)
```
